## 4_Merging Performance and Survey outcomes

```python
import pandas as pd
import numpy as np
```

### 1. Input 2 outcomes data into dataframe

```python
data1 = pd.read_stata("PerformanceOutcomes_Randomization_CPP_Merged.dta")
data2 = pd.read_stata("SurveyOutcomes_Randomization_CPP_Merged.dta")
```

```python
data1.head()
```

### 2. Update missing values in PerformanceOutcomes from SurveyOutcomes

```python
data1.update(data2, overwrite=False)
data1
```

### 3. Merge two outcomes data 

```python
outcome_merged = pd.merge(data1, data2, on=['email', 'time'], how='outer')
outcome_merged = outcome_merged[:371]
outcome_merged
```

### 4. Generating correct RA for baseline balance checks

```python
# gen col coach_same
for i in range(len(outcome_merged)):
    outcome_merged.loc[i, "coach_same"] = [ 1 if outcome_merged.loc[i, "fall2018coachingra_x"] == 
                                           outcome_merged.loc[i, "spring2019coachingra_x"] else 0]
```

```python
# gen col spring2019coachingra_correct
for i in range(len(outcome_merged)):
    if outcome_merged.loc[i, "fall2018coachingra_x"]==1 and outcome_merged.loc[i, "coach_same"]==1:
        outcome_merged.loc[i, "spring2019coachingra_correct"] = 1
    elif outcome_merged.loc[i, "fall2018coachingra_x"]==0 and outcome_merged.loc[i, "coach_same"]==1:
        outcome_merged.loc[i, "spring2019coachingra_correct"] = 0 
    elif outcome_merged.loc[i, "coach_same"]==0:
        outcome_merged.loc[i, "spring2019coachingra_correct"] = outcome_merged.loc[i, "spring2019coachingra_x"] 
```

```python
# gen col coach_correct_same     
for i in range(len(outcome_merged)):
        outcome_merged.loc[i, "coach_correct_same"] = [ 1 if outcome_merged.loc[i, "fall2018coachingra_x"] == 
                                                       outcome_merged.loc[i, "spring2019coachingra_correct"] else 0]
```

### 8.  Generating variables with missing values

```python
def colMissingVals(df, columnList):
    ''' Generate a binary column that indicates whether the given column has a missing value.
    Provide the dataframe and the column of interest. The output will be the dataframe with one 
    additional column named {columnName}_miss that has a 1 for missing values in columnName
    and a 0 for a non-missing value in columnName.
    '''
    for columnName in columnList:
        newcol = columnName + "_miss"   # create a missing column name
        for i in range(len(df)):
            df.loc[i, newcol] = [1 if pd.isnull(df.loc[i, columnName]) else 0]
    return df
```

```python
missing_cols_list = ["ccs_gpa", "partch_either", "moedu_colab", "faedu_colab", "gender_female", "race_white", 
                     "age_21ab", "hsses_1", "hsses_2", "hsses_3", "hsrace_1", "hsrace_2", "hsrace_3", "hsach_1", 
                     "hsach_2", "hsach_3", "hsloc_1", "hsloc_2", "hsloc_3"]
colMissingVals(outcome_merged_sifted, missing_cols_list)
```
