## Functions

```python
def createTableQuery(all_column_list, char_col_list, tableName, primaryKeys):
    query = "CREATE TABLE " + tableName + " ("
    for col in all_column_list:
        if col in char_col_list:
            query = query + col + " varchar(100), "
        else:
            query = query + col + " float, "
    for key in primaryKeys:
        query = query + " PRIMARY KEY (" + key + "), "
    query = query[:-2] + " );"
    return query
```

```python
def createTableQuery2(all_column_list, char_col_list, big_char_list, tableName, primaryKeys):
    query = "CREATE TABLE " + tableName + " ("
    for col in all_column_list:
        if col in big_char_list:
            query = query + col + " varchar(10000), "
        elif col in char_col_list:
            query = query + col + " varchar(100), "
        else:
            query = query + col + " float, "
    for key in primaryKeys:
        query = query + " PRIMARY KEY (" + key + "), "
    query = query[:-2] + " );"
    return query
```

```python
def showTable(tableName, cursor):
    query = "SELECT * FROM " + tableName + ";"
    cursor.execute(query)
    result = cursor.fetchall()
    colnames = [x[0] for x in cursor.description]
    df = pd.DataFrame(result, columns=colnames)
    return df
```
